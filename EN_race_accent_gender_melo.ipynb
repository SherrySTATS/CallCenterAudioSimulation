```python
import re

def clean_text(text):
    # Remove content between parentheses
    text = re.sub(r'\(.*?\)', '', text)
    # Remove "Customer:" and "Agent:" including any leading/trailing spaces
    cleaned_text = re.sub(r'\b(Customer:|Agent:)\s*', '', text, flags=re.IGNORECASE)
    # Replace newlines with a space
    cleaned_text = cleaned_text.replace("\n", " ").strip()
    # Remove extra spaces
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text)
    return cleaned_text
```

```python
# Parameters

speed = 1.0  # Adjust speed if necessary
device = 'cpu'  # 'cuda' for GPU, 'cpu' for CPU

def clean_text(text):
    """ Remove 'Customer:', 'Agent:', content between parentheses, and all numbers """
    # Remove content within parentheses
    text = re.sub(r'\(.*?\)', '', text)
    # Remove 'Customer:' and 'Agent:'
    text = re.sub(r'\b(Customer:|Agent:)\s*', '', text, flags=re.IGNORECASE)
    # Remove all numbers
    text = re.sub(r'\d+', '', text)
    # Replace newlines with space and remove extra whitespace
    text = text.replace("\n", " ").strip()
    text = re.sub(r'\s+', ' ', text)
    return text


# Fixed parameters
language_index = "0"  # 0 for English Transcripts
text_index = "01"
race_test_index = "01"
prefix = "2222222222"
accent_index = 5 # Chinese accent

filename_list = []

# Loop through all transcripts
for row_index in range(len(transcripts)):
    # Format row index as two digits
    row_index_str = f"{row_index + 1:02d}"

    # Clean the text
    text = clean_text(transcripts.Transcript[row_index])

    # Initialize model for Chinese
    model = TTS(language='ZH', device=device)
    speaker_ids = model.hps.data.spk2id

    # Construct the filename core
    filename_core = f"{language_index}{text_index}{race_test_index}{accent_index}{row_index_str}"

    # Construct the full filename with 'record_' prefix
    output_filename = f"{prefix}{filename_core}_{prefix}{filename_core}.wav"
    output_path = f"{output_path_main}{output_filename}"

    filename_list.append(output_filename)

    # Generate the audio file
    model.tts_to_file(text, speaker_ids['ZH'], output_path, speed=speed)
```

```python
# Save the record to an Excel file
output_excel_path = output_path_main+"en_fairness_race_record_filenames.xlsx"
record_df['EN-ZH'] = pd.Series(filename_list)
record_df.to_excel(output_excel_path, index=False)

print(f"File record saved to {output_excel_path}")
```

```python
import re
import random
import os
from TTS.api import TTS

# Parameters
output_path_main = "/content/drive/My Drive/text-to-speech/gender/EN_gender_audio/"  # Update this path
speed = 1.0  # Adjust speed if necessary
#speaker
male = ['Luis Moray','Marcos Rudaski','Ferran Simen','Eugenio Mataracı','Kumar Dahl','Aaron Dreschner']
female = ['Daisy Studious','Gracie Wise','Brenda Stern','Sofia Hellen','Gitta Nikolina','Asya Anara']

# Fixed parameters
language_index = "0" #0+：English， 1=:Spanish
text_index = "01" #01=:fairness test
gender_test_index = "02" #02=:gender test
prefix = "2222222222"

# Gender indices and keys
gender_indices = [0, 1] #0=:female, 1=:male
genders = ['female','male']
#genders = ['male']

# Dataframe to record filenames
columns = ["Row Index"] + genders
record_df_gender = pd.DataFrame(columns=columns)

# Use gpu=True if a GPU is available and configured
#get device
device = "cuda" if torch.cuda.is_available() else "cpu"
# Initialize model
tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2").to(device)

# Loop through all transcripts
for row_index in range(len(transcript)):
    # Format row index as two digits
    row_index_str = f"{row_index + 1:02d}"

    # Clean the text
    text_clean = transcript.tts_transcripts[row_index]
    #text_clean = "El conocimiento es poder, pero compartirlo es sabiduría."

    # Record filenames for each gender
    filenames_gender = [row_index_str]

    for gender_index, gender in enumerate(genders):
        filename_core = f"{language_index}{text_index}{gender_test_index}{gender_index}{row_index_str}"
        # Ref audio
        # random choose one speaker from the gender
        if gender == 'male':
          spearker_name = random.choice(male)
        else:
          spearker_name = random.choice(female)

        # Construct the full filename
        output_filename = f"{prefix}{filename_core}_{prefix}{filename_core}.wav"
        output_path = f"{output_path_main}{output_filename}"
        #print(output_filename)

        # Generate the audio file
        tts.tts_to_file(text=text_clean,
                file_path=output_path,
                speaker=spearker_name,
                language="en")

        # Append filename to the list
        filenames_gender.append(output_filename)

    # Add the row to the dataframe
    record_df_gender.loc[len(record_df_gender)] = filenames_gender

```

```python
# Save the record to an Excel file
output_excel_path = output_path_main+"EN_record_df_gender_filenames.xlsx"

record_df_gender.to_excel(output_excel_path, index=False)

print(f"File record saved to {output_excel_path}")
```
