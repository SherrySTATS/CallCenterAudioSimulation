```python
import re

def clean_text(text):
    # Remove content between parentheses
    text = re.sub(r'\(.*?\)', '', text)
    # Remove "Customer:" and "Agent:" including any leading/trailing spaces
    cleaned_text = re.sub(r'\b(Cliente:|Agente:)\s*', '', text, flags=re.IGNORECASE)
    # Replace newlines with a space
    cleaned_text = cleaned_text.replace("\n", " ").strip()
    # Remove extra spaces
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text)
    return cleaned_text
```

### generate audio based on the accent reference audio

```python
import re
import random
import os

# Parameters
output_path_main = "/content/drive/My Drive/text-to-speech/race/ES_race_audio/"  # Update this path
speed = 1.0  # Adjust speed if necessary


# Fixed parameters
language_index = "1" #1=:Spanish
text_index = "01"
race_test_index = "01"
prefix = "2222222222"

# Accent indices and keys
accent_indices = [0, 1, 2, 3, 4,5]
accents = ['argentinian', 'chilean', 'colombian', 'peruvian', 'Puerto_Rican','Venezuelan']

# Dataframe to record filenames
columns = ["Row Index"] + accents
record_df = pd.DataFrame(columns=columns)

# Use gpu=True if a GPU is available and configured
#get device
device = "cuda" if torch.cuda.is_available() else "cpu"
# Initialize model
tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2").to(device)

# Loop through all transcripts
for row_index in range(len(transcript)):
    # Format row index as two digits
    row_index_str = f"{row_index + 1:02d}"

    # Clean the text
    text_clean = transcript.tts_transcripts[row_index]

    # Record filenames for each accent
    filenames = [row_index_str]

    for accent_index, accent in enumerate(accents):
        filename_core = f"{language_index}{text_index}{race_test_index}{accent_index}{row_index_str}"
        # Ref audio
        # random choose one reference from the accent file
        ref_accent_path = ref_audio_dir+accent+'/'
        ref_audio_files = [f for f in os.listdir(ref_accent_path) if f.endswith('.wav')]
        ref_file =random.choice(ref_audio_files)
        ref_audio_path = os.path.join(ref_accent_path,ref_file)

        # Construct the full filename
        output_filename = f"{prefix}{filename_core}_{prefix}{filename_core}.wav"
        output_path = f"{output_path_main}{output_filename}"

        # Generate the audio file
        tts.tts_to_file(text=text_clean,
                file_path=output_path,
                speaker_wav=ref_audio_path,
                language="es")

        # Append filename to the list
        filenames.append(output_filename)

    # Add the row to the dataframe
    record_df.loc[len(record_df)] = filenames

```

```python
# Save the record to an Excel file
output_excel_path = output_path_main+"es_fairness_race_record_filenames.xlsx"

record_df.to_excel(output_excel_path, index=False)

print(f"File record saved to {output_excel_path}")
```
